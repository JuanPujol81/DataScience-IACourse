The coefficient of variation is a standardized measure of the dispersion of a probability distribution or dataset. It is calculated as the ratio of the standard deviation to the mean, expressed as a percentage:

CV=(σ / μ)×100%

Where σ is the standar deviation and μ in the mean.


```python
# Data
data_A = [70, 72, 75, 78, 80]
data_B = [55, 65, 75, 85, 95]

# Desviación estándar
std_A = np.std(data_A, ddof=1)
std_B = np.std(data_B, ddof=1)

# Mean
mean_A = np.mean(data_A)
mean_B = np.mean(data_B)

# Coefficient of variation.
cv_A = (std_A / mean_A) * 100
cv_B = (std_B / mean_B) * 100

print(f"Clase A - Standar Deviation: {std_A}, CV: {cv_A}%")
print(f"Clase B - Standar Deviation: {std_B}, CV: {cv_B}%")
```
