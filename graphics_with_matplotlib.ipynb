```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("Crecimiento de ventas")
plt.xlabel("Días")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()
```

```python
productos = ['A', 'B', 'C', 'D']
ventas = [200, 150, 300, 100]

plt.bar(productos, ventas, color='green')
plt.title("Ventas por producto")
plt.xlabel("Producto")
plt.ylabel("Unidades vendidas")
plt.show()
```

```python
edades = [25, 30, 35, 25, 40, 45, 50, 35, 30, 30, 25]

plt.hist(edades, bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
import pandas as pd

# Creamos un DataFrame de ejemplo
df = pd.DataFrame({
    'Edad': [25, 30, 45, 22, 35, 50, 41],
    'Salario': [30000, 40000, 50000, 25000, 42000, 60000, 52000]
    })

# Gráfico de dispersión
sns.scatterplot(x='Edad', y='Salario', data=df)
plt.title("Relación entre Edad y Salario")
plt.show()
```
