```python
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
import pandas as pd

# Creamos un DataFrame de ejemplo
df = pd.DataFrame({
    'Edad': [25, 30, 45, 22, 35, 50, 41],
    'Salario': [30000, 40000, 50000, 25000, 42000, 60000, 52000]
    })

# GrÃ¡fico de dispersiÃ³n
sns.scatterplot(x='Edad', y='Salario', data=df)
plt.title("RelaciÃ³n entre Edad y Salario")
plt.show()
```

```python
# Boxplot por departamento
empleados = pd.DataFrame({
    'Departamento': ['Ventas', 'Marketing', 'TI', 'TI', 'Ventas', 'Marketing', 'Ventas'],
    'Salario': [50000, 60000, 70000, 62000, 55000, 70000, 50000]
})

sns.boxplot(x='Departamento', y='Salario', data=empleados)
plt.title("DistribuciÃ³n de salario por departamento")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
data = {
    'Name': ['Ana', 'Luis', 'Carlos', 'Pedro', 'LucÃ­a', 'Ricardo', 'Julia'],
    'Age': [25, 30, 32, 45, 29, 41, 38],
    'Salary': [50000, 60000, 70000, 62000, 58000, 70000, 67000],
    'Department': ['Ventas', 'Marketing', 'TI', 'TI', 'Ventas', 'Marketing', 'Ventas']
    }

empleados = pd.DataFrame(data)

# Matriz de correlaciÃ³n
correlation = empleados.select_dtypes(include=['number']).corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("CorrelaciÃ³n entre variables")
plt.show()
```

### Matriz de correlaciÃ³n
El cÃ³digo genera una matriz de correlaciÃ³n para analizar la relaciÃ³n entre las variables numÃ©ricas de un DataFrame. La funciÃ³n `df.corr()` calcula los coeficientes de correlaciÃ³n entre las columnas numÃ©ricas del DataFrame.

Luego, se utiliza `sns.heatmap` para visualizar esta matriz como un mapa de calor. El argumento `annot=True` muestra los valores numÃ©ricos de la correlaciÃ³n dentro de cada celda, y `cmap='coolwarm'` define el esquema de colores para el mapa de calor.

Finalmente, `plt.title` agrega un tÃ­tulo al grÃ¡fico, y `plt.show()` muestra el grÃ¡fico generado.

### Â¿QuÃ© es la correlaciÃ³n entre variables?
La correlaciÃ³n mide la relaciÃ³n entre dos variables y cÃ³mo cambian juntas. Se expresa como un valor entre -1 y 1:

- **1** indica una correlaciÃ³n positiva perfecta: cuando una variable aumenta, la otra tambiÃ©n aumenta proporcionalmente.
- **0** indica que no hay correlaciÃ³n: las variables no tienen relaciÃ³n lineal.
- **-1** indica una correlaciÃ³n negativa perfecta: cuando una variable aumenta, la otra disminuye proporcionalmente.

La correlaciÃ³n se utiliza para identificar patrones y relaciones entre variables en un conjunto de datos, lo que puede ser Ãºtil para anÃ¡lisis predictivos y estadÃ­sticos.
