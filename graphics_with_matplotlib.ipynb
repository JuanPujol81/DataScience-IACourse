```python
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
import pandas as pd

# Creamos un DataFrame de ejemplo
df = pd.DataFrame({
    'Edad': [25, 30, 45, 22, 35, 50, 41],
    'Salario': [30000, 40000, 50000, 25000, 42000, 60000, 52000]
    })

# Gráfico de dispersión
sns.scatterplot(x='Edad', y='Salario', data=df)
plt.title("Relación entre Edad y Salario")
plt.show()
```

```python
# Boxplot por departamento
empleados = pd.DataFrame({
    'Departamento': ['Ventas', 'Marketing', 'TI', 'TI', 'Ventas', 'Marketing', 'Ventas'],
    'Salario': [50000, 60000, 70000, 62000, 55000, 70000, 50000]
})

sns.boxplot(x='Departamento', y='Salario', data=empleados)
plt.title("Distribución de salario por departamento")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
data = {
    'Name': ['Ana', 'Luis', 'Carlos', 'Pedro', 'Lucía', 'Ricardo', 'Julia'],
    'Age': [25, 30, 32, 45, 29, 41, 38],
    'Salary': [50000, 60000, 70000, 62000, 58000, 70000, 67000],
    'Department': ['Ventas', 'Marketing', 'TI', 'TI', 'Ventas', 'Marketing', 'Ventas']
    }

empleados = pd.DataFrame(data)

# Matriz de correlación
correlation = empleados.select_dtypes(include=['number']).corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlación entre variables")
plt.show()
```

### Matriz de correlación
El código genera una matriz de correlación para analizar la relación entre las variables numéricas de un DataFrame. La función `df.corr()` calcula los coeficientes de correlación entre las columnas numéricas del DataFrame.

Luego, se utiliza `sns.heatmap` para visualizar esta matriz como un mapa de calor. El argumento `annot=True` muestra los valores numéricos de la correlación dentro de cada celda, y `cmap='coolwarm'` define el esquema de colores para el mapa de calor.

Finalmente, `plt.title` agrega un título al gráfico, y `plt.show()` muestra el gráfico generado.

### ¿Qué es la correlación entre variables?
La correlación mide la relación entre dos variables y cómo cambian juntas. Se expresa como un valor entre -1 y 1:

- **1** indica una correlación positiva perfecta: cuando una variable aumenta, la otra también aumenta proporcionalmente.
- **0** indica que no hay correlación: las variables no tienen relación lineal.
- **-1** indica una correlación negativa perfecta: cuando una variable aumenta, la otra disminuye proporcionalmente.

La correlación se utiliza para identificar patrones y relaciones entre variables en un conjunto de datos, lo que puede ser útil para análisis predictivos y estadísticos.

```python
# Ya pandas, seaborn y matplotlib están importados, podemos crear un DataFrame de ejemplo
empleados_tarea = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Pedro', 'Marta', 'Lucía', 'Jorge'],
    'Edad': [25, 32, 45, 28, 38, 50, 41],
    'Salario': [50000, 62000, 70000, 54000, 65000, 72000, 68000],
    'Departamento': ['Ventas', 'Marketing', 'TI', 'TI', 'Ventas', 'Marketing', 'TI']
    })

```

```python
# Gráfico de dispersión
sns.scatterplot(x ='Edad', y ='Salario', data = empleados_tarea)
plt.title("Relación entre Edad y Salario")
plt.show()
```

```python
sns.boxplot(x ='Departamento', y ='Salario', data = empleados_tarea)
plt.title("Distribución de salario por departamento")
plt.show()
```

```python
correlation = empleados_tarea.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlación entre variables")
plt.show()
```
