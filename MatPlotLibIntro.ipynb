MAtPlotLib examples workflow.

```python
# 1-Prepare data:
my_x = [1, 2, 3, 4]
my_y = [11, 22, 33, 44]

# 2-Setup plot:
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3-Plot data:
ax.plot(my_x, my_y)

# 4-Customize plot:
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5-Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

Making figure with numpy arrays:
- Line plot
- Scatter plot
- Bar plot
- Histogram
- Subplots

```python
# Create a line plot by default.
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot.
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Make a plot from dictionary.
peanut_butter_prices = {"Almond Butter": 10,
                        "Peanut Butter": 8,
                        "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(peanut_butter_prices.keys(), peanut_butter_prices.values())
ax.set(title="Butter Store", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(peanut_butter_prices.keys()), list(peanut_butter_prices.values()));
ax.set(title="Butter Store", xlabel="Price ($)");
```

```python
# Make some data for histograms and plot it.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

Two options for plotting data:

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(peanut_butter_prices.keys(), peanut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(peanut_butter_prices.keys(), peanut_butter_prices.values())    
ax[1, 1].hist(np.random.randn(1000));
```


Plotting from pandas data frames.

```python
car_sales = pd.read_csv("car-sales.csv")
```

```python
car_sales.head()
```

```python
car_sales.replace({"Price": '[\$\,]'}, '', regex=True, inplace=True)
car_sales["Price"] = car_sales["Price"].astype(float)
car_sales
```

```python
ts = pd.Series(car_sales.Price.values, index=car_sales.Make.values)

ts = ts.cumsum()

ts.plot();
```
