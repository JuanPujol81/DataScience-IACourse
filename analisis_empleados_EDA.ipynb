Analisis de Datos.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print(np.array([1, 2, 3]))        # Debería funcionar
sns.set_theme(style="whitegrid")                         # Debería configurar el estilo de Seaborn
plt.plot([1, 2, 3], [4, 5, 6])    # Debería dibujar una línea
plt.show()
```

```python
emplojees = pd.read_excel("./data_set/Employees.xlsx", usecols=["First Name", "Years", "Monthly Salary", "Department", "Country"])  # Cargar el archivo Excel
# Cargar el archivo Excel
print(emplojees.head())           # Mostrar las primeras filas del DataFrame
```

```python
print(emplojees.count())        # Mostrar estadísticas descriptivas del DataFrame
```

```python
print("Median")
print(emplojees.median(numeric_only=True))  # Mostrar la mediana de las columnas numéricas
print("Mean")
print(emplojees.mean(numeric_only=True))          # Mostrar la media de las columnas numéricas
print("std")
print(emplojees.std(numeric_only=True))            # Mostrar la desviación estándar de las columnas numéricas
```

```python
emplojees.drop_duplicates(inplace=True)  # Eliminar duplicados
emplojees.dropna(inplace=True)  # Eliminar filas con valores NaN
emplojees.reset_index(drop=True, inplace=True)  # Reiniciar el índice
print(emplojees.isnull().count())       # Mostrar estadísticas descriptivas del DataFrame
```

```python
print(emplojees.describe())
```


Visualizaciones:

Gráfico de dispersión entre Edad y Salario.

Boxplot de Edad por Departamento.

Heatmap de correlaciones entre variables numéricas.


```python
sns.set_theme(style="whitegrid")
sns.scatterplot(data=emplojees, x="Years", y="Monthly Salary", hue="Department", style="Country")
plt.title("Scatter plot of Monthly Salary vs Years")
plt.show()  # Mostrar el gráfico
# Mostrar el gráfico de dispersión
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=emplojees, x="Department", y="Monthly Salary")
plt.title("Box plot of Monthly Salary by Department")
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x
plt.tight_layout
plt.show()  # Mostrar el gráfico
```

```python
# Matriz de correlación
correlation = emplojees.corr(numeric_only= True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlación entre variables")
plt.show()
```

Analisis de Resultados:
De los datos recopilados podemos ver que los anos de servicio no tienen u a relacion directa con el salario de forma general, los salarios estan mas relacionados con el departamento a que pertenece el empleado. De esto podemos concluir que es una empresa con alta rotacion de personal, que ademas basa sus ofertas laborales en la renovacion de los perfiles profesionales.
